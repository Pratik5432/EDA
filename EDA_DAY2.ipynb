### 3. Group Analysis:
* What are the average measurements for each species (mean sepal length, sepal width, petal length, and petal width)?
* Create a bar chart showing the average petal length for each species.

```python
df.groupby('species')[['sepal_length',	'sepal_width', 'petal_length',	'petal_width']].mean()
```

```python
df.groupby('species')[['sepal_length',	'sepal_width', 'petal_length',	'petal_width']].mean().plot(kind='bar')
```

### 4. Correlation Analysis:

```python
df.select_dtypes(include='number').corr()
```

```python
sns.heatmap(df.select_dtypes(include='number').corr())
```

### 5. Data Visualization:
